```python
data_dir = "/HPS/oraclase/work/reproduction/images/"
all_rgb_colors = np.array(Image.open(f"{data_dir}/result.png"))
all_rgb_colors = all_rgb_colors.reshape(-1, 3)
```

```python
class GamutMapping:
    def __init__(self, all_rgb_colors):
        self.all_rgb_colors = all_rgb_colors
        self.gm_background = np.array(Image.open("/HPS/oraclase/work/reproduction/images/gm_backround.png"))
        self.paste_region_corners = np.array([[ 906, 1021],
                            [ 915, 1979],
                            [1985, 1021],
                            [1997, 1982]])
        self.paste_region_xmin, self.paste_region_ymin = self.paste_region_corners.min(0)
        self.gm_background_size = (self.paste_region_corners.max(0) - self.paste_region_corners.min(0))[::-1] 
        
    def gamut_mapping(self, input_img):        
        '''
        input_img = user provided WxHx3/4 image 
        all_rgb_colors = pre-saved look-up table
        returns gamut mapped image
        '''
        has_alpha_channel = input_img.shape[-1]==4
        if has_alpha_channel:
            alpha_channel = input_img[:, :, -1:]
            input_img = input_img[:, :, :3]
        idx = input_img[:, :, 0]*256**2 + input_img[:, :, 1]*256 + input_img[:, :, 2]
        gm_out = self.all_rgb_colors[idx].reshape(input_img.shape).reshape(input_img.shape)

        scaler = self.gm_background_size[0]/max(gm_out.shape[:2])
        gm_image_size = (int(scaler*gm_out.shape[1]), int(scaler*gm_out.shape[0]))
        gm_out_resized = cv2.resize(gm_out, gm_image_size)
        offset = (self.gm_background_size - gm_image_size)/2
        xmin, ymin = int(self.paste_region_xmin + offset[1]), int(self.paste_region_ymin + offset[0])

        gm_background = np.copy(self.gm_background)
        # fill the original image
        gm_background[xmin:xmin+gm_image_size[1], ymin:ymin+gm_image_size[0]] = gm_out_resized

        if has_alpha_channel:    
            # pdb.set_trace()
            gm_out = np.concatenate([gm_out, alpha_channel], axis=-1)
        return gm_out, gm_background
```

```python
gm = GamutMapping(all_rgb_colors)
```

```python
input_img = np.array(Image.open(f"{data_dir}/ladywall.jpg"))
```

```python
%%timeit
gm_out, gm_with_background = gm.gamut_mapping(input_img)
```
